Q1 What is the difference between multithreading and multiprocessing?
Ans Multithreading and Multiprocessing are two distinct concurrency approaches in Python that differ fundamentally in how they achieve parallelism, memory usage, and execution model.
1.Core definition - Multithreading runs multiple threads within a single process, sharing the same memory space. Multiprocessing runs multiple processes each with its own independent memory space.
2.Global Interpreter Lock (GIL) - Python's GIL is the critical distinction - it allows only one thread to execute Python bytecode at a time, making true parallelism impossible with threads. Multiprocessing bypasses the GIL entirely by using separate processes.
3.True parallelism - Multithreading achieves concurrency but not true parallelism in CPU-bound tasks due to the GIL. Multiprocessing achieves true parallelism by utilizing multiple CPU cores simultaneously.
4.Memory model - Threads share memory space - all threads within a process access the same variables and data structures. Processes have completely isolated memory - data must be explicitly passed between them via IPC mechanisms.
5.Communication - Threads communicate directly via shared memory, which is fast but requires synchronization. Processes communicate via Inter-Process Communication (IPC) - pipes, queues, and shared memory objects.
6.Overhead - Thread creation is lightweight and fast - threads share process resources. Process creation is heavyweight and slow - each process duplicates the parent's memory space via forking.
7.Best use case for threading - Multithreading excels at I/O-bound tasks - file reading, network requests, database calls - where threads wait for external resources and the GIL is released during waiting.
8.Best use case for multiprocessing - Multiprocessing excels at CPU-bound tasks - data processing, mathematical computation, image rendering - where actual parallel CPU execution is required.
9.Crash isolation - A crashed thread can bring down the entire process and all sibling threads. A crashed process is fully isolated - it does not affect the parent or sibling processes.
10.Synchronization complexity - Threads require careful synchronization via locks, semaphores, and events to prevent race conditions on shared data. Processes avoid this since memory is isolated, but require explicit data serialization for communication.
11.Scalability - Multithreading is limited by the GIL and does not scale well across multiple CPU cores for CPU-bound work. Multiprocessing scales linearly with the number of available CPU cores.
12.Python modules - Multithreading uses the threading module. Multiprocessing uses the multiprocessing module, which provides a nearly identical API for easier transition between the two.
13.Debugging difficulty - Multithreading bugs like race conditions and deadlocks are notoriously hard to reproduce and debug. Multiprocessing bugs are typically more deterministic due to memory isolation.
14.Context switching - Thread context switching is handled by the OS scheduler within a single process, making it faster. Process context switching involves full OS-level switching, which is significantly more expensive.

Q2 What are the challenges associated with memory management in Python?
Ans Challenges in Memory Management in Python:-
Automatic Management: Python handles memory itself, so developers have limited direct control.
Garbage Collection Limits: Cyclic references may not be freed immediately.
Memory Leaks: Poor handling of objects (e.g., global references) can retain unused memory.
High Overhead: Python objects consume more memory compared to low-level languages.
Fragmentation: Memory may become scattered, reducing efficient reuse.
Reference Counting Issues: Frequent updates to reference counts add overhead.
Large Data Handling: Managing big datasets can lead to high memory usage.

Q7 Explain the difference between apply() and map() in Pandas
Ans apply() and map() are two distinct Pandas methods for transforming data, differing fundamentally in scope, flexibility, and intended use case.
Scope of operation - map() operates element-wise on a single Series only. apply() operates on entire rows or columns of a DataFrame, or element-wise on a Series.
Primary target - map() is designed exclusively for Series objects. apply() works on both Series and DataFrames, making it far more versatile.
Input it passes to function - map() passes individual scalar values to the function one at a time. apply() passes entire rows or columns as Series objects by default.
Function complexity - map() is suited for simple, scalar transformations - lookups, replacements, basic formatting. apply() handles complex, multi-column logic requiring access to entire rows.
Dictionary and Series support - map() uniquely accepts dictionaries and Series as direct mappers for value substitution, without needing a function at all. apply() strictly requires a callable.
Performance - map() is faster for simple element-wise transformations as it has less overhead. apply() is slower due to row/column-level operation and Python-level looping.
axis parameter - apply() supports the axis parameter - axis=0 for column-wise, axis=1 for row-wise operations. map() has no axis concept - it always works element by element.
NaN handling - map() propagates NaN values silently by default when using dictionary mapping - missing keys return NaN. apply() passes NaN directly to the function for explicit handling.
Return type flexibility - apply() can return scalars, Series, or DataFrames depending on the function used. map() always returns a Series of the same length.
Ideal use case - Use map() for simple value replacement or formatting on a single column. Use apply() when logic spans multiple columns or requires complex row-level computation.
